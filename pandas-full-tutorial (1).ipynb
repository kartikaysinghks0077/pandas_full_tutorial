```python
import pandas as pd

data = {'A': [10, 20, 20, 30, 40]}
df = pd.DataFrame(data)

# Default ranking (method='average', ascending=True)
df['rank_default'] = df['A'].rank()

# Rank with method='min'
df['rank_min'] = df['A'].rank(method='min')

# Rank with method='max'
df['rank_max'] = df['A'].rank(method='max')

# Rank with method='first'
df['rank_first'] = df['A'].rank(method='first')

# Rank with method='dense'
df['rank_dense'] = df['A'].rank(method='dense')

# Rank in descending order
df['rank_desc'] = df['A'].rank(ascending=False)

print(df)
```

```python
data = {'A': [1, 2, 3, 4, 5]}
df = pd.DataFrame(data)

# Rolling sum with a window size of 3
df['rolling_sum_3'] = df['A'].rolling(window=3).sum()

# Rolling mean with a window size of 3
df['rolling_mean_3'] = df['A'].rolling(window=3).mean()

# Rolling sum with a window size of 3 and minimum periods of 1
df['rolling_sum_3_min1'] = df['A'].rolling(window=3, min_periods=1).sum()

# Rolling sum with a window size of 3, centered
df['rolling_sum_3_center'] = df['A'].rolling(window=3, center=True).sum()

print(df)
```

```python
import pandas as pd

# Sample DataFrame
data = {'born_country': ['USA', 'FRA', 'GER', 'HUN', 'ITA', 'HND', None]}
bios = pd.DataFrame(data)

# Convert to lowercase and uppercase
bios['born_country_lower'] = bios['born_country'].str.lower()
bios['born_country_upper'] = bios['born_country'].str.upper()

# Get string length
bios['country_length'] = bios['born_country'].str.len()

# Replace substring
bios['born_country_replaced'] = bios['born_country'].str.replace('USA', 'United States')

# Extract first and last two characters
bios['first_two_chars'] = bios['born_country'].str[:2]
bios['last_two_chars'] = bios['born_country'].str[-2:]

# Strip whitespace
bios['born_country_stripped'] = bios['born_country'].str.strip()

# Split strings (assuming some countries are written with spaces)
# For the sake of demonstration, let's update one value
bios.loc[0, 'born_country'] = 'United States of America'
bios[['part1', 'part2', 'part3', 'part4']] = bios['born_country'].str.split(' ', expand=True, n=3)

# Display the DataFrame
print(bios)
```

```python
import pandas as pd

# Sample DataFrame
data = {'born_country': ['USA', 'FRA', 'GER', 'HUN', 'ITA']}
bios = pd.DataFrame(data)

# Method 1: Assign a constant value
bios['constant_column'] = 'constant_value'

# Method 2: Assign values from a list
bios['list_column'] = [1, 2, 3, 4, 5]

# Method 3: Assign values using a function
bios['length_column'] = bios['born_country'].apply(len)

# Method 4: Add a column based on conditions
bios['size_category'] = bios['born_country'].apply(lambda x: 'Short' if len(x) < 4 else 'Long')

# Method 5: Using assign() method
bios = bios.assign(assign_column=[10, 20, 30, 40, 50])

print(bios)
```

```python
import pandas as pd

# Sample DataFrame
data = {'born_date': ['2024-08-07', '1990-12-25', '1985-05-14']}
bio = pd.DataFrame(data)

# Convert 'born_date' column to datetime
bio['born_date'] = pd.to_datetime(bio['born_date'])

print(bio)
```

```python
# Extract year, month, and day
bio['year'] = bio['born_date'].dt.year
bio['month'] = bio['born_date'].dt.month
bio['day'] = bio['born_date'].dt.day

print(bio)
```

```python
# Extract day of the week
bio['day_of_week'] = bio['born_date'].dt.day_name()

# Extract quarter
bio['quarter'] = bio['born_date'].dt.quarter

# Extract week of the year
bio['week_of_year'] = bio['born_date'].dt.isocalendar().week

# Example DataFrame with time
data_time = {'timestamp': ['2024-08-07 14:23:45', '1990-12-25 09:15:30', '1985-05-14 22:30:00']}
bio_time = pd.DataFrame(data_time)

# Convert to datetime
bio_time['timestamp'] = pd.to_datetime(bio_time['timestamp'])

# Extract hour, minute, second
bio_time['hour'] = bio_time['timestamp'].dt.hour
bio_time['minute'] = bio_time['timestamp'].dt.minute
bio_time['second'] = bio_time['timestamp'].dt.second

print(bio_time)
```
